```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('whole_pic4.png')
template = cv2.imread('TooCon.png')
_, w, h = template.shape[::-1]

res = cv2.matchTemplate(img,template,cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
top_left = max_loc
btm_right = (top_left[0] + w, top_left[1] + h)

cv2.rectangle(img,top_left, btm_right, 255, 2)

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```
import cv2
import os

def save_all_frames(video_path, dir_path, basename, ext='jpg'):
    cap=cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print('wrong')
    else:
        print('good')

    os.makedirs(dir_path, exist_ok=True)
    base_path = os.path.join(dir_path, basename)

    digit = len(str(int(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))/30)))
    frames=int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    n = 0
    while True:
        ret, frame = cap.read()
        if ret and n%30==0:
            cv2.imwrite('{}_{}.{}'.format(base_path, str(int(n/30)).zfill(digit), ext), frame)

        if n>frames:
            break
        n+=1

save_all_frames('/content/data_wrong/1-3.mp4', '/content/image_wrong/', 'Wrong_1-3', 'png')
```
